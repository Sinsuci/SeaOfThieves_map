```python
import pyMeow as pm
import pymem
import time
from math import *


def initialisation():
    pm.overlay_init()
    fps = 20
    pm.set_fps(fps)

    color_white = pm.get_color("white")
    color_red = pm.get_color("red")
    map_full = pm.load_texture("simple.png")

    offset_menu_x = 5
    offset_menu_y = 5
    show_menu = False
    print("close : END \n Show num_0 \n Hide num_6")
    return color_white, color_red, map_full, offset_menu_x, offset_menu_y, show_menu, fps

def menu_status(show_menu):
    # if pm.key_pressed(0x60) :
    #     if show_menu == False :
    #         show_menu = True
    #     else :
    #         show_menu = False
    #     time.sleep(0.3) #0.3
    # return show_menu
    if pm.key_pressed(0x60) :
        show_menu = True
    if pm.key_pressed(0x66) :
        show_menu = False
    return show_menu


def draw_pointer(x_pointer, y_pointer,player_dir, color):
    pointer_thickness = 20
    pm.draw_pixel(x_pointer, y_pointer, color)
    pm.draw_line(x_pointer-pointer_thickness/2, y_pointer, x_pointer+pointer_thickness/2, y_pointer, color, 1)
    pm.draw_line(x_pointer, y_pointer-pointer_thickness/2, x_pointer, y_pointer+pointer_thickness/2, color, 1)
    pm.draw_circle_lines(x_pointer, y_pointer, pointer_thickness/2, color)
    angle = player_dir/180*pi
    pm.draw_line(x_pointer, y_pointer, x_pointer+cos(angle)*pointer_thickness, y_pointer+sin(angle)*pointer_thickness, color, 1)

def grab_data(process, oplayer):
    player_x = process.read_float(oplayer+0x78)
    player_y = process.read_float(oplayer+0x7C)
    player_dir = process.read_float(oplayer+0x88)
    return player_x, player_y, player_dir

def pixel_to_map(x_map, y_map):
    x_pixel = 0.000921*x_map+562
    y_pixel = 0.000921*y_map+539
    return x_pixel, y_pixel


color_white, color_red, map_full, offset_menu_x, offset_menu_y, show_menu, fps = initialisation()

pid = pm.get_process_id('SoTGame.exe')
process = pymem.Pymem()
process.open_process_from_id(pid)


def get_player_data(process):
    u_world_base = 0x7F81E1
    WorldOwningGameInstance = 0x1C0
    GameInstanceLocalPlayers = 0x38
    u_offset = process.read_ulong(process.base_address + u_world_base + 0x3)
    oworld = process.read_longlong(process.base_address + u_world_base + u_offset + 0x7)
    game_instance = process.read_longlong(oworld + WorldOwningGameInstance)
    players = process.read_longlong(game_instance + GameInstanceLocalPlayers)
    oplayer = process.read_longlong(players)
    return oplayer

oplayer = get_player_data(process)

while pm.overlay_loop():
    # start = time.time()
    pm.begin_drawing()
    show_menu = menu_status(show_menu)  # change variable affichage menue


    if show_menu == True :
        zoom = 0.38
        pm.draw_texture(map_full, offset_menu_x, offset_menu_y, color_white, 0, zoom)
        pm.draw_rectangle_lines(offset_menu_x, offset_menu_y, map_full["width"]*zoom, map_full["height"]*zoom, color_white, 1)

        player_x, player_y, player_dir = grab_data(process, oplayer)
        x_pixel, y_pixel = pixel_to_map(player_x, player_y)
        draw_pointer(x_pixel, y_pixel, player_dir, color_red)
        time.sleep(0.3)

    pm.end_drawing()
    # if time.time()-start < 1/fps:
    #     time.sleep(1/fps-(time.time()-start))

```
